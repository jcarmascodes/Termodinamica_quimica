A continuación se describe el procedimiento para la obtención del coeficiente de actividad para la obtención de un diagrama $P-xy$ con su respectivo diagrama $x-y$.


1. Primero se calculan los volúmenes moleculares relativas ($r_i$) y las áreas moleculares relativas ($q_i$) individuales para cada grupo funcional presentes en las moléculas, utilizando tanto los parámetros de volumen funcional relativa ($R_k$) y el de área funcional relativa ($Q_k$) que corresponden a cada grupo funcional (tomando en cuenta cuales grupos funcionales están presentes en cada compuesto de la mezcla, utilizando los parámetros $\nu_{k}^{(i)}$ que es el parámetro de frecuencia de grupos funcionales para cada compuesto).


\begin{equation}
    r_{i}\:=\:\sum_{k}\:\nu_{k}^{(i)}R_{k}
\end{equation}


\begin{equation}
    q_{i}\:=\:\sum_{k}\:\nu_{k}^{(i)}Q_{k}
\end{equation}


Por ejemplo, siguiendo el ejercicio las ecuaciones (2 compuestos, 5 grupos funcionales individuales involucrados) quedarían de la siguiente manera:


\begin{equation}
    r_{1}\:=\:\nu_{1}^{(1)}R_{1}\:+\:\nu_{2}^{(1)}R_{2}\:+\:\nu_{3}^{(1)}R_{3}\:+\:\nu_{4}^{(1)}R_{4}\:+\:\nu_{5}^{(1)}R_{5}
\end{equation}


\begin{equation}
    r_{2}\:=\:\nu_{1}^{(2)}R_{1}\:+\:\nu_{2}^{(2)}R_{2}\:+\:\nu_{3}^{(2)}R_{3}\:+\:\nu_{4}^{(2)}R_{4}\:+\:\nu_{5}^{(2)}R_{5}
\end{equation}


\begin{equation}
    q_{1}\:=\:\nu_{1}^{(1)}Q_{1}\:+\:\nu_{2}^{(1)}Q_{2}\:+\:\nu_{3}^{(1)}Q_{3}\:+\:\nu_{4}^{(1)}Q_{4}\:+\:\nu_{5}^{(1)}Q_{5}
\end{equation}


\begin{equation}
    q_{2}\:=\:\nu_{1}^{(2)}Q_{1}\:+\:\nu_{2}^{(2)}Q_{2}\:+\:\nu_{3}^{(2)}Q_{3}\:+\:\nu_{4}^{(2)}Q_{4}\:+\:\nu_{5}^{(2)}Q_{5}
\end{equation}


$\newline$


2. Luego se calculan las áreas parciales de los grupos funcionales ($e_ki$), se utiliza el área funcional relativa ($Q_k$) que corresponden a cada grupo funcional con respecto a las moléculas presentes a la mezcla junto con el parámetro $\nu_{k}^{(i)}$ que es el parámetro de frecuencia de grupos funcionales para cada compuesto y por último las áreas moleculares relativas ($q_i$) de cada molécula. Recordando que los subíndices corresponden al número de grupo funcional correspondiente y el superíndices le corresponden a la molécula presente en la molécula. Entonces la ecuación correspondiente para obtener $e_ki$, sería la siguiente.


\begin{equation}
    e_{ki}\:=\:\frac{\nu_{k}^{(i)}Q_{k}}{q_{i}}
\end{equation}


$\newline$


3. Después se realiza el cálculo de las energías funcionales por interacción de grupos funcionales ($\tau_{mk}$) en donde este parámetro está en función de la temperatura de la mezcla ($T$, $K$) junto con los parámetros de energía de interacción entre los diferentes grupos funcionales ($a_{mk}$), se utiliza la siguiente ecuación en donde tanto los subíndices $m$ y $k$ son los diferentes grupos funcionales individuales.


\begin{equation}
    \tau_{mk}\:=\:e^{\dfrac{-a_{mk}}{T}}
\end{equation}


$\newline$


4. Posteriormente, se realiza el cálculo de la energía parcial por grupo funcional en cada especie química ($\beta_{ik}$), en donde se utiliza sumatoria entre el producto el área parcial de los grupos funcionales ($e_{ki}$) de cada especie química y cada una de las energías funcionales por interacción ($\tau_{mk}$) entre su propio grupo funcional y entre otras que estén presente entre las especies químicas presentes en la mezcla. En este caso los subíndices es por cada especie química presente en la mezcla y los subíndices m y k son de los correspondientes grupos funcionales; entonces la ecuación queda de la siguiente manera.


\begin{equation}
    \beta_{ik}\:=\:\sum_{m}\:e_{mi}\tau_{mk}
\end{equation}


$\newline$


5. Luego, se procede a calcular el área fraccional por cada grupo funcional ($\theta_{k}$), en donde se utiliza cada fracción de cada componente de la mezcla ($x_{i}$ y $x_{j}$) junto con las áreas moleculares relativas ($q_i$ y $q_j$) individuales para cada grupo funcional presentes en cada molécula con sus correspondientes áreas parciales de grupos funcionales por cada molécula ($e_ki$).


\begin{equation}
    \theta_{k}\:=\:\dfrac{\sum_{i}\:x_{i}q_{i}e_{ki}}{\sum_{j}\:x_{j}q_{j}}
\end{equation}


$\newline$

```python
#CÓDIGO INCOMPLETO - EN PRUEBAS Y DESARROLLO DE FUNCIONAMIENTO#

# - CALCULAR EL VOLUMEN MOLECULAR RELATIVO (r_i) Y EL ÁREA MOLECULAR RELATIVA (q_i)
ri=[]
qi=[]
for j in range(len(VkT)):
    N_COMP_ri=[]
    N_COMP_qi=[]
    for i in range(len(Vk1)):
        N_COMP_ri.append(RkS[i]*(VkT[j][i]))
        N_COMP_qi.append(QkS[i]*(VkT[j][i]))
    ri.append(N_COMP_ri)
    qi.append(N_COMP_qi)
print(ri)
print(qi)

ri_TOTAL=[]
qi_TOTAL=[]
for i in ri:
    ri_TOTAL.append(sum(i))
for i in qi:
    qi_TOTAL.append(sum(i))
print(ri_TOTAL)
print(qi_TOTAL)

# - CALCULAR EL ÁREA PARCIAL DE GRUPO FUNCIONAL e_ki
e_ki=[]
for j in range(len(VkT)):
    N_COMP_eki=[]
    for i in range(len(Vk1)):
        N_COMP_eki.append((QkS[i]*(VkT[j][i]))/qi_TOTAL[j])
    e_ki.append(N_COMP_eki)
print(e_ki)

# - CALCULAR LA ENERGÍA FUNCIONAL POR INTERACCIÓN DE GRUPOS FUNCIONALES T_mk
T_mk=[]
for j in range(len(GFC)):
    N_GF_Tmk=[]
    for i in range(len(GFI_1)):
        N_GF_Tmk.append(m.exp(-GFC[j][i]/T_SIST))
    T_mk.append(N_GF_Tmk)
print(T_mk)

# - CALCULAR LA ENERGÍA PARCIAL POR GRUPO FUNCIONAL EN CADA ESPECIE (beta)_ik
beta_ik = [[0 for _ in range(len(e_ki))] for _ in range(5)]
for i in range(5):
    COMP_SUM = [0] * len(e_ki)
    for j in range(5):
        for k in range(len(e_ki)):
            COMP_SUM[k] += e_ki[k][j] * T_mk[j][i]
    for k in range(len(e_ki)):
        beta_ik[i][k] = COMP_SUM[k]
print(beta_ik)

# - GENERACIÓN Y CALCULO DE LAS FRACCIONES MOL DE CADA COMPONENTE
X1=[]
X2=[]
n=5
for i in range(n):
    X1.append(0.25*i)
    X2.append(1-X1[i])
X=[X1,X2]

# - CALCULAR EL ÁREA FRACCIONAL POR GRUPO FUNCIONAL POR CADA ESPECIE QUÍMICA Y POR CADA GRUPO FUNCIONAL (theta)_k
theta_k = [[0 for _ in range(len(X1))] for _ in range(5)]
for i in range(5):
    for j in range(n):
        numerator = ((X[0][j] * qi_TOTAL[0] * e_ki[0][i]) + (X[1][j] * qi_TOTAL[1] * e_ki[1][i]))
        denominator = ((X[0][j] * qi_TOTAL[0]) + (X[1][j] * qi_TOTAL[1]))
        theta_k[i][j] = numerator / denominator 
print(theta_k)  
        
# - CALCULAR LA ENERGÍA FRACCIONAL POR GRUPO FUNCIONAL
```

A continuación se describe el procedimiento para la obtención del coeficiente de actividad para la obtención de un diagrama $\;T-xy\;$ con su respectivo diagrama $\;x-y\;$. Sin embargo, el procedimiento cambia con respecto al apartado anterior ya que se obtiene la temperatura en donde se obtiene la presión total de la mezcla a la cantidad que se fija como objetivo, utilizando el método de convergencia por Newton-Raphson; con el siguiente procedimiento:


a. Primero se calcula los pasos 1 al 6 se calcula a partir de una temperatura supuesta ($T_{sup}$) (en este caso se puede comenzar en cualquier temperatura).


b. Luego, se calcula un cambio a la temperatura sumpuesta ($T_{sup}'$) (puede utilizar un $\Delta T=1$ o cualquier otra cantidad para obtener un cierto grado de presición de los datos obtenidos).


\begin{equation}
    T_{sup}'\:=\:T_{sup}\:+\:\Delta T
\end{equation}


c. Luego, se vuelve a calcular los pasos 1 al 6 con la nueva temperatura supuesta $T_{sup}'$


d. Posteriormente, se calcula el margen de error de la presión obtenida con las temperaturas supuestas con la presión objetivo. En cuanto obtener E (calculando $p_{calc}$ con $T_{sup}$) tanto E' (calculando $p_{calc}'$ con $T_{sup}'$), se utiliza las siguientes ecuaciones:


\begin{equation}
    E\:=\:E'\:=\:\ln\left(\frac{p_{calc}}{p}\right)
\end{equation}


e. Luego, se obtiene una nueva temperatura supuesta ($T_{Nueva}$) con la ecuación de convergencia de Newton-Raphson propuesta para este caso.


\begin{equation}
    T_{Nueva}\;=\;\frac{T_{sup}T_{sup}'\left(E'\;-\;E\right)}{T_{sup}'E'\;-\;T_{sup}E}
\end{equation}


f. Posteiormente, se repetirá el proceso de los incisos A a la E hasta que el error de la (E con respecto a $T_{sup}$) tenga un resultado igual o menor a la tolerancia marcada en el programa


g. Por último, ya obtenido las presiones parciales de cada componente y la temperatura para alcanzar la presión objetivo se procede con el paso 7 para obtener su correspondiente fracción mol del gas.


h. Para graficar el diagrama $T-xy$ se debe de repetir los incisos a hasta g con cada una de las fracciones mol en fase líquido entre los dos compuestos, siguiendo el paso 8.
